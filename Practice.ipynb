```python
def sep(p):
    li=[]
    a=""
    for u in p:
        if u!="/":
            a+=u
        else:
            li.append(a)
            a=""
    li.append(a)
    td2=int(li[0])
    tm2=int(li[1])
    ty2=int(li[2])
    lio=td2,tm2,ty2
    return lio
months=[31,28,31,30,31,30,31,31,30,31,30,31]
Todaysdate=input("Enter Todays date in format dd/mm/yy i.e 12/1/2020 : ")
we=sep(Todaysdate)
Birthdate=input("Enter Your Date of Birth in format dd/mm/yy i.e 20/2/2001 : ")
ew=sep(Birthdate)
ay=we[2]-ew[2]-1
am=we[1]-1+12-ew[1]
ad=we[0]+months[ew[1]-1]-ew[0]
if ad>=months[ew[1]-1]:
    am+=1
    ad-=months[ew[1]-1]
if am>=12:
    ay+=1
    am-=12
if am==ad==0:
    print(f"Happy Birthday! You are {ay} years old today.")
else:
    print(f"Your Age is {ay} years {am} months {ad} days")
```

```python
#Program to convert a binary number into a decimal number
def btdc(abinary):
    #abinary=input("Enter a Binary number")
    y=0
    decimalnumber=0
    bl=[]
    for j in abinary:
        bl.append(int(j))
    bl.reverse()
    for i in bl:
        p=power(2,y)*i
        decimalnumber+=p
        y+=1
    return (decimalnumber)
```

```python
#Function to solve quadratic formulae
def qf(a,b,c):
    import math
    det=b**2-4*a*c
    if det==0:
        print("Roots are :\nx1 = %.2f\nx2 = %.2f"%((-b/(2*a)),(-b/(2*a))))
    elif det>0:
        print("Roots are :\nx1 = %.2f\nx2 = %.2f"%(((-b+math.sqrt(det))/(2*a)),((-b-math.sqrt(det))/(2*a))))
    elif det<0:
        print("Roots are :\nx1 = %.2f + %.2fi\nx2 = %.2f + %.2fi"%(-b/2*a,(math.sqrt(-1*det))/(2*a),-b/2*a,(math.sqrt(-1*det))/(2*a)))
```
