```python
import numpy as np
myarray=np.array([[1,-3,5,2],[6,7,1,8],[3,3,12,5]])
a=np.zeros((1,4))
for i in range (0,4):
    sum=0
    for j in range(0,3):
        sum=sum+myarray[j][i]
    a[0][i]=sum
mynewarray=np.vstack((myarray,a))
def ExtractDiag(a):
    b=np.zeros((1,4))
    for i in range(0,4):
        for j in range(0,4):
            if i==j:
                b[0][i]=a[i][j]
    return b
diagArrayReturned=ExtractDiag(mynewarray)
diagArrayReturned
diagArrayReturned[0]=0
diagArrayReturned[1]=10
diagArrayReturned[2]=15
diagArrayReturned[3]=100
```
